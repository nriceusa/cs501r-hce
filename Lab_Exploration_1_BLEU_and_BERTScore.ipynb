BLEU is an automatic metric, originally developed for machine translation evaluation, which has been used widely in NLG evaluation. The basic idea of BLEU is that it compares matching word sequences (n-grams) between target and generated text. For a quick rundown, you may wish to view [this video](https://www.youtube.com/watch?v=M05L1DhFqcw).

BERTScore is also an automatic metric, but instead of relying on n-grams, which is a purely syntactic approach, it measures semantic similarity between target and generated text. It does so using embeddings. If you are unfamiliar with embeddings, you may wish to spend some time learning about them through [this article](https://huggingface.co/spaces/hesamation/primer-llm-embedding?section=what_are_embeddings?) and other resources.
